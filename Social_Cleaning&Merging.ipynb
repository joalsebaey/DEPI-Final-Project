# Cleaning & Merging the data for better EDA

### Importing & Loading What We Need

```python
import pandas as pd

# Load DateDimension.csv
data_df = pd.read_csv('Social/DateDimension.csv')

# Load PostDimension.csv
post_df = pd.read_csv('Social/PostDimension.csv')

# Load SentimentFact.csv
fact_df = pd.read_csv('Social/SentimentFact.csv')

# Load UserDimension.csv
user_df = pd.read_csv('Social/UserDimension.csv')
```

### Merging The Data Into One File

```python
# Fact_df & Data_df
sentiment_fact = pd.merge(fact_df,data_df, on = 'DateID')
```

```python
sentiment_fact
```

```python
# sentiment_fact & post_df
sentiment_fact = pd.merge(sentiment_fact,post_df, on = 'PostID')
```

```python
sentiment_fact = pd.merge(sentiment_fact, user_df, on='UserID')
```

### Saving Our Data

```python
sentiment_fact.to_csv('sentiment_factDEPI.csv', index=False)
```
